clustering by the colors with violet in the color profile

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Define the predefined colors
color_names = ['Yellow', 'Orange', 'Red', 'Violet', 'Blue', 'Green', 'Black', 'White']

# Load the CSV file
csv_file_path = r'C:\Users\yesha\Desktop\College\4th year\1st Semester\THS-ST1\Color_Extraction\downscaling_1571.csv'
df = pd.read_csv(csv_file_path)

# Remove any rows with missing values or incorrect formats
df = df.dropna()

# Extract the color percentages (ignoring the 'Name' column)
color_data = df.iloc[:, 1:].astype(float).values

# Choose the number of clusters (experiment with different values)
num_clusters = 8  # You can change this based on your preference

# Apply K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(color_data)

# Display cluster assignments
print(df[['Name', 'Cluster']])

# Visualizing clusters in a 2D space (projecting using PCA)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(color_data)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Color Clustering of Images using K-Means')
plt.colorbar(scatter, label='Cluster')
plt.show()
```
